Tirando 3 informações de 25 lojas de um shopping:
- Faturamento de cada Loja;
- Quantidade de produtos vendidos de cada Loja;
- Ticket médio dos produto de cada Loja.

```python
import pandas as pd
```

```python
dados = pd.read_excel('vendas.xlsx')
dados
```

## Quantidade de lojas 

```python
dados['ID Loja'].value_counts()
```

# Faturamento de cada Loja

```python
faturamento_por_loja = dados[["ID Loja", "Valor Final"]].groupby("ID Loja").sum()
faturamento_por_loja = faturamento_por_loja.sort_values(by='Valor Final', ascending=False)
faturamento_por_loja
```

# Quantidade de produtos vendidos de cada Loja

```python
produtos_por_loja = dados[["ID Loja", "Quantidade"]].groupby("ID Loja").sum()
produtos_por_loja = produtos_por_loja.sort_values(by='Quantidade', ascending=False)

produtos_por_loja
```

# Ticket médio dos produto de cada Loja

```python
ticket_medio = (faturamento_por_loja['Valor Final'] / produtos_por_loja['Quantidade']).to_frame('Ticket Médio')
ticket_medio = ticket_medio.sort_values(by='Ticket Médio', ascending=False)

ticket_medio
```
